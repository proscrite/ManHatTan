### Let's start supposing the plural form and the dativ declination are correct, but punctuation and (obviously) capitalization don't coincide

```python
myansw = 'randern'  # This would be the input
```

```python
answ == myansw
```

```python
accented_string.lower(), answ.lower()
```

```python
import unicodedata
accented_string = u'Málaga'
# accented_string is of type 'unicode'
def strip_accents(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')
filtansw = strip_accents(answ.lower())
filtansw
```

```python
filtansw == myansw
```
