# Create evolution trees

```python
def read_pkmn_datafile(datapath) -> pd.DataFrame:

    with open(datapath, "r") as f:
        data = f.read()

    # Split the data into sections for each Pokémon
    entries = data.split("[")
    entries = [e for e in entries if e.strip()]  # Remove empty sections

    # Parse each Pokémon entry
    pokemon_list = []
    for entry in entries:
        lines = entry.splitlines()
        pokemon_data = {}
        for line in lines:
            if "=" in line:
                key, value = line.split("=", 1)  # Split into key and value
                pokemon_data[key] = value
        pokemon_list.append(pokemon_data)

    # Convert the list of dictionaries into a Pandas DataFrame
    df = pd.DataFrame(pokemon_list)
    df.dropna(how='all', inplace=True)
    # Display the DataFrame
    return df
```

```python
datapath = '/Users/pabloherrero/Downloads/POKEMON ANIL V3.06/JUEGO/PBS/pokemon.txt'

pkdf = read_pkmn_datafile(datapath)
```

```python
# Parse the "Evolutions" column and build relationships
evolution_edges = []

for _, row in pkdf.iterrows():
    if pd.notna(row["Evolutions"]):  # Ensure "Evolutions" is not NaN
        evolutions = row["Evolutions"].split(",")  # Split evolution string
        for i in range(0, len(evolutions), 3):  # Process in steps of 3
            next_evolution = evolutions[i]
            # Add edge from current Pokémon to its evolution
            evolution_edges.append((row["InternalName"], next_evolution))

# Create a directed graph
G = nx.DiGraph()
G.add_edges_from(evolution_edges)
from collections import defaultdict

# Group Pokémon by evolution stage
stages = defaultdict(list)


stages = defaultdict(int)

# Traverse the graph to determine node depth
for node in nx.topological_sort(G):  # Topological sort ensures parents appear before children
    predecessors = list(G.predecessors(node))
    stage = max([stages[pred] for pred in predecessors], default=-1) + 1 if predecessors else 0
    stages[node] = stage
```

```python
stage_0_pokemon = [pokemon for pokemon, stage in stages.items() if stage == 0]
stage_0_pokemon[:4]
```

```python
index_stage_0 = pkdf[pkdf['InternalName'].isin(stage_0_pokemon)].index
index_stage_0_df = pd.DataFrame(index_stage_0, columns=["Index"])

pout = '/Users/pabloherrero/Documents/ManHatTan/gui/Graphics/index_stage_0.csv'
index_stage_0_df.to_csv(pout, index=False)
```
